Seaborn

Importar y traer datos de ejemplo

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset de ejemplo
tips = sns.load_dataset("tips")

# Mostrar las primeras filas
print(tips.head())
```

Gráfico de dispersión (scatter plot)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

Histograma

```python
sns.histplot(data=tips, x="total_bill", bins=20)
plt.show()
```

Boxplot (diagrama de caja)

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

Gráfico de barras

```python
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
```

Gráfico de violín (violin plot)

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.show()
```

Añadir título y etiquetas a los ejes

Después de crear el gráfico con Seaborn, usa:

    plt.title("Título del gráfico") para el título

    plt.xlabel("Etiqueta eje X") para la etiqueta del eje X

    plt.ylabel("Etiqueta eje Y") para la etiqueta del eje Y

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Relación entre total de la cuenta y la propina")
plt.xlabel("Total de la cuenta ($)")
plt.ylabel("Propina ($)")
plt.show()
```

Añadir texto adicional o anotaciones

Para poner textos dentro del gráfico en un punto específico, usa:

plt.text(x, y, "Tu texto aquí", fontsize=12, color="red")

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Relación entre total de la cuenta y la propina")
plt.xlabel("Total de la cuenta ($)")
plt.ylabel("Propina ($)")
plt.text(40, 8, "Aquí las propinas más altas", fontsize=12, color="blue")
plt.show()
```
